### Central Tendancy
- Average
    - The single most value to represent our data
    - Need not to be data point itself
    - Considers every point in our data
    - Discrete as well as continous
    - It is vulnerable to outliers
- Median
    - Value such that 50% of the data on the other side
    - More robust to outliers in the dataset
    - Does not take care of every datapoint in the dataset
- Mode
    - most frequent value in the dataset
    - the highest bar in the histogram
    - typicaly used in categorical data
- Geometric Mean
    - Used for sumarizing ratios such as the compound anual growth rate
    
- Harmonic mean
    - Great for sumarizing rates such as resistors in parallel or P/E ratios in finance

- Range
- Interquartile

- Mean deviation
    - a measure of how much each data point varies from the mean
    - squared mean deviation is the square of the mean deviation
- Variance 
    - the mean of the sum of squared mean deviation 
    
- rather than use the variance directly, we get the standard deviation which is the square root of the variance
