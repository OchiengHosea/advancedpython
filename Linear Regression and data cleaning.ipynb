```python
import pandas as pd
auto_data = pd.read_csv('data/imports-85.data', sep=r'\s*,\s*', engine="python")
auto_data
```

```python
import numpy as np
auto_data = auto_data.replace('?', np.nan)
auto_data.head()
```

```python
auto_data.describe()
```

```python
auto_data.describe(include="all")
```

```python
auto_data['price'] = pd.to_numeric(auto_data['price'], errors='coerce')
auto_data['price'].describe()
```

```python
auto_data = auto_data.drop('normalized-losses', axis=1)
auto_data.head()
```

```python
auto_data.describe(include='all')
```

```python
auto_data['horsepower'].describe()
```

```python
auto_data['horse_power'] = pd.to_numeric(auto_data['horsepower'], errors='coerce')
```

```python
cylinders_dict = {'two':2, 'three': 3, 'four':4, 'five': 5, 'six':6, 'eight':8, 'twelve':12}
auto_data['num-of-cylinders'].replace(cylinders_dict, inplace=True)
auto_data.head()
```

```python
auto_data = pd.get_dummies(auto_data, columns=['make', 'fuel-type', 'aspiration', 'num-of-doors', 'body-style', 'drive-wheels', 'engine-location', 'engine-type', 'fuel-type', 'fuel-system'])
auto_data.head()    
```

```python
auto_data[auto_data.isnull().any(axis=1)]
```

```python
auto_data = auto_data.dropna()
```

```python
auto_data[auto_data.isnull().any(axis=1)]
auto_data.to_csv('data/cleaned_imports.csv')
```

```python
from sklearn.model_selection import train_test_split
X = auto_data.drop('price', axis=1)
Y = auto_data['price']

X_train, x_test, Y_train, y_test = train_test_split(X, Y, test_size=0.2, random_state=0)
```

```python
from sklearn.linear_model import LinearRegression

linear_model = LinearRegression()
linear_model.fit(X_train, Y_train)
```

```python
linear_model.score(X_train, Y_train)
```

```python
predictors = X_train.columns
coef = pd.Series(linear_model.coef_, predictors).sort_values()
```
