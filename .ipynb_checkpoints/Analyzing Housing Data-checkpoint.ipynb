### Backward Difference Encoder

This technique falls under the contrast coding system for categorical features. A feature of K categories, or levels, usually enters a regression as a sequence of K-1 dummy variables. In backward difference coding, the mean of the dependent variable for a level is compared with the mean of the dependent variable for the prior level. This type of coding may be useful for a nominal or an ordinal variable.

If you want to learn other contrast coding methods you can check out this resource.

The code structure is pretty much the same as any method in the category_encoders library, just this time you will call BackwardDifferenceEncoder from it:

Having looked at various methods of encoding categorical variable, lets do a detailed EDA of the Categorical values first befor we decide on the strategies of cleaning them up and wrangling them

```python
# change categorical column type from object to category, this makes operations on the fields to be a little faster during preprocessing 
# and building the model, you can check using the timeit magic function 
non_numeric_fields = non_numeric_fields.astype('category')
total_obs = len(non_numeric_fields)
# Define helper functions to help us describe our data

def get_percentage(x, y):
    """Gets the percentage between two numeric figures"""
    return ((y - x) / y) * 100

def get_mode_report(var, dtype='categorical'):
    """gets the mode of a categorical variable, the frequency and percentage occurrence in the column"""
    if(dtype=='categorical'):
        data_frame = non_numeric_fields
    else:
        data_frame = numeric_fields
    mode = data_frame[var].mode()
    freq = data_frame[var].value_counts()[mode][0]
    pct = get_percentage(freq, data_frame[var].shape[0])
    return f'{mode} occurs {freq} times, this is about {pct: .2f} % of {mode}'

is_na_count = lambda var, dtype='categorical': non_numeric_fields[var][non_numeric_fields[var].isna()].shape[0] if dtype=='categorical' else numeric_fields[var][numeric_fields[var].isna()].shape[0]
is_na_pct = lambda var, dtype='categorical': round(get_percentage(non_numeric_fields[var][non_numeric_fields[var].isna()].shape[0], total_obs), 2) if dtype=='categorical' else round(get_percentage(numeric_fields[var][numeric_fields[var].isna()].shape[0], total_obs), 2)
unique_categs_report = lambda var: f'{len(non_numeric_fields[var].unique())} Unique values where {get_mode_report(var)}'

def run_na_on_categs(dtype='categorical'):
    if dtype == 'categorical':
        data_frame = non_numeric_fields
    else:
        data_frame = numeric_fields
        
    return pd.DataFrame({var: 
             {
                 'na_cnt': is_na_count(var, dtype),
                 'pct_complete': is_na_pct(var, dtype)
    #                  'unique_values': unique_report(var)
             }
     for var in data_frame.columns.tolist()})
na_df = run_na_on_categs()
na_df
```

```python
# Replace Na in Alley with NoAl
# replace NA in BSMt with NB
# replace NAs BSMt1, BSMt2 with NB1, NB2
# non_numeric_fields['Alley'] = non_numeric_fields['Alley'].cat.add_categories('NoAl')
replace_na_dicts = {
    'BsmtQual': 'NB',
    'BsmtCond': 'NB',
    'BsmtFinType1': 'NB',
    'BsmtFinType2': 'NB',
}
for key in replace_na_dicts.keys():
    if replace_na_dicts[key] not in list(non_numeric_fields[key].astype('category').cat.categories.tolist()):
        non_numeric_fields[key] = non_numeric_fields[key].cat.add_categories(replace_na_dicts[key])
    non_numeric_fields[key].fillna(replace_na_dicts[key], inplace=True)
non_numeric_fields.head()
```

### Further Cleaning of Categorical variables
We can see that there are still some observations having NA after rectifications

Now we will adopt the following criteria for completely cleaning our categorical variables

- Fill all NA with the mode if NA is less than 30%
- If variable is of significance to SalePrice and NA is more than 30%, replace NA in the variable with the mode of the variable that it is highly correlated with apart from SalesPrice, the variable must be categorical
- Discard the variable if it has no correlation with SalePrice(i.e its less than 0.5 or greator than -0.5) and its NA is more than 30%

```python
# select variables that are not 100% complete and are less than 70% complete
def get_cols_to_purge(upper_bound, lower_bound, dtype='categorical'):
    if dtype == 'categorical':
        data_frame = na_df
    else:
        data_frame = num_na_df
    return [{col:data_frame[col]['pct_complete']} for col in data_frame.columns
        if data_frame[col]['pct_complete'] < upper_bound and data_frame[col]['pct_complete'] < lower_bound and col != 'SalePrice']
# call the function and specify 100 and 70 as bounds
get_cols_to_purge(100, 70)
```

Thus we can see that we have 3 variables to clean further, and FirepLaceQu needs to be filled by the mode since its NA values are less than 30 %
we will write a function to do this automatically for any other columns

```python
def fill_mode(cols_na_pctgs, dtype='categorical'):
    if dtype == 'categorical':
        data_frame = non_numeric_fields
    else:
        data_frame = numeric_fields
    for col in cols_na_pctgs:
        key = list(col.keys())[0]
        if col[key] > 30:
            data_frame[key].fillna(data_frame[key].mode()[0], inplace=True)
            
# lets call the function on our list of variables
fill_mode(get_cols_to_purge(100, 70))
```
