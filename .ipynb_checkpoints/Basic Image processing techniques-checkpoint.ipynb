```python
from skimage import io
from skimage import color, data, exposure, transform
from skimage.measure import compare_ssim as ssim
from pylab import *
import pandas as pd
img = io.imread('boot.jpg')
df = pd.DataFrame(img.flatten())
filepath = 'pixel_values1.xlsx'
df.to_excel(filepath, index=False)
# io.imshow(img)
```

```python
img.shape
```

```python
# converting colorspace
# convert to hsv
img_hsv = color.rgb2hsv(img)
img_rgb = color.hsv2rgb(img_hsv)
figure(0)
io.imshow(img_hsv)
figure(1)
io.imshow(img_rgb)
```

```python
# Saving an image
# convert to YPbPr
img_ypbpr = color.rgb2ypbpr(img)
# convert back to rgb
img_rgb = color.ypbpr2rgb(img_ypbpr)
io.imsave('boot_ypbpr.jpg', img_ypbpr)
```

```python
figure(0)
io.imshow(img_ypbpr)
```

```python
# Gamma correction
# we use exposure class that contains a function called adjust_gamma, which we use to 
# give an image as an input and the final gamma value that we want
gamma_corrected1 = exposure.adjust_gamma(img, 0.5)
gamma_corrected2 = exposure.adjust_gamma(img, 5)
figure(0)
io.imshow(gamma_corrected1)
figure(1)
io.imshow(gamma_corrected2)
```

```python
# rotating, shifting and scaling images
img_rot = transform.rotate(img, 20)
io.imshow(img_rot)
```

```python
# resize
img_res = resize(img, (100, 100))
io.imshow(img_res)
io.imsave('ss.jpg', img_res)
```

```python
# determining structural similarity
img_modified = io.imread('boot_mod.jpg')
ssim_original = ssim(img, img, data_range=img.max() - img.min(), multichannel = True)
ssim_different = ssim(img, img_modified, data_range=img_modified.max()-img_modified.min(), multichannel=True)
print(ssim_original, ssim_different)
```
